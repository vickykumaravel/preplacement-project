# In which year did the player hit the most 4s and 6s combined?

```python
odi1['Total_4s_6s'] = odi1['4s'] + odi1['6s']
max_year = odi1.loc[odi1['Total_4s_6s'].idxmax(), 'Year']
print(max_year)
```

```python
plt.bar(odi1['Year'], odi1['Total_4s_6s'], color='blue', alpha=0.7, label='4s + 6s')
plt.xlabel('Year')
plt.ylabel('Total 4s and 6s Combined')
plt.title('Total 4s and 6s Combined Over the Years')
plt.xticks(rotation=45)
plt.legend()
plt.show()
```

# What percentage of the player's runs come from boundaries (4s and 6s)

```python
total_runs = 9205  

# Total number of 4s and 6s
total_4s = 834
total_6s = 234

# Calculate the total runs from boundaries
runs_from_boundaries = total_4s * 4 + total_6s * 6

# Calculate the percentage
percentage_from_boundaries = (runs_from_boundaries / total_runs) * 100

print(percentage_from_boundaries)
```

```python
labels = ['Runs from Boundaries', 'Other Runs']
sizes = [percentage_from_boundaries, 100 - percentage_from_boundaries]
colors = ['lightcoral', 'lightskyblue']
plt.pie(sizes,labels=labels, colors=colors, autopct='%1.1f%%', shadow=True, startangle=140) 
plt.title('Percentage of Runs from Boundaries')
plt.show()
```

```python
# From the above graph we came to know that 51.5% has come from boundries and without boundries he has scored 48.5%.
```
